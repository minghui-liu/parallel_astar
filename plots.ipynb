```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

### Sequential Timing with Varying K values

```python
K_list = [4, 8, 16, 32]

seq_timing = [12894, 18747, 18588, 6353]
```

```python
plt.plot(K_list, seq_timing, marker='o')
#plt.plot(num_gpus, charmpp, marker='o', label='Charm++')
plt.title("Sequential A* execution time v/s K value")
plt.xlabel("K value")
plt.ylabel("Time (ms)")
```

### Charm++ vs MPI

```python
num_procs = [1, 2, 4, 8, 16, 32]

mpi = [14686, 7667, 4182, 2381, 1474, 1457]
charmpp = [15615, 27487.91, 13151.05, 10339.19, 8747.55, 8020.69]
```

```python
plt.plot(num_procs, mpi, marker='o', label='MPI')
plt.plot(num_procs, charmpp, marker='o', label='Charm++')
plt.title("Execution Time v/s Number of Processes")
plt.legend()
plt.xlabel("Number of Processes")
plt.ylabel("Time (ms)")
```

### Experiment 3

```python
num_procs = [1, 2, 4, 8, 16, 32]

node_1000 = [10, 8, 9, 6, 16, 122]
node_100000 = [1350, 725, 441, 262, 304, 275]
node_1000000 = [14804, 7744, 4231, 2477, 1434, 1093]

plt.plot(num_procs, node_1000, marker='o', label='1000 Nodes')
plt.plot(num_procs, node_100000, marker='o', label='100000 Nodes')
plt.plot(num_procs, node_1000000, marker='o', label='1000000 Nodes')
plt.title("Execution Time v/s Number of Processes")
plt.legend()
plt.xlabel("Number of Processes")
plt.ylabel("Time (ms)")
```

### Emphasizing the importance of termination algorithms 

```python
num_procs = [1, 2, 4, 8, 16, 32]

barrier = [9947, 5245, 2911, 1786, 8611, 19510]
modified_barrier = [9994, 5277, 2927, 1611, 642, 553]

plt.plot(num_procs, barrier, marker='o', label='Barrier')
plt.plot(num_procs, modified_barrier, marker='o', label='Modified Barrier')
plt.title("Execution Time v/s Number of Processes")
plt.legend()
plt.xlabel("Number of Processes")
plt.ylabel("Time (ms)")
```
