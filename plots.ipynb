### Speedup

```python
def calc_speedup(arr):
    speedup = []
    for time in arr:
        speedup.append(arr[0]/time)
    return speedup
```

```python
def calc_efficiency(arr):
    speedup = []
    i = 0
    for time in arr:
        speedup.append(arr[0]/(time*2**i))
        i += 1
    return speedup
```

```python
num_procs = [1, 2, 4, 8, 16, 32]

node_1000 = [10, 8, 9, 6, 16, 122]
node_100000 = [1350, 725, 441, 262, 304, 275]
node_1000000 = [14804, 7744, 4231, 2477, 1434, 1093]

speedup_1000 = calc_speedup(node_1000)
speedup_100000 = calc_speedup(node_100000)
speedup_1000000 = calc_speedup(node_1000000)

plt.plot(num_procs, speedup_1000, marker='o', label='1000 Nodes')
plt.plot(num_procs, speedup_100000, marker='o', label='100000 Nodes')
plt.plot(num_procs, speedup_1000000, marker='o', label='1000000 Nodes')
plt.title("Speedup v/s Number of Processes")
plt.legend()
plt.xlabel("Number of Processes")
plt.ylabel("Speedup")
```

### Efficiency

```python
efficiency_1000 = calc_efficiency(node_1000)
efficiency_100000 = calc_efficiency(node_100000)
efficiency_1000000 = calc_efficiency(node_1000000)

plt.plot(num_procs, efficiency_1000, marker='o', label='1000 Nodes')
plt.plot(num_procs, efficiency_100000, marker='o', label='100000 Nodes')
plt.plot(num_procs, efficiency_1000000, marker='o', label='1000000 Nodes')
plt.title("Efficiency v/s Number of Processes")
plt.legend()
plt.xlabel("Number of Processes")
plt.ylabel("Efficiency")
```
